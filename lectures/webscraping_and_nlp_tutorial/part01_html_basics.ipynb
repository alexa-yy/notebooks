## HTML

__HTML__ is a markup language, its basic build blocks are the `<tag>`s.<br>
(Almost) every `<tag>` has two parts:

- Opening `<tag>` 
- Closing `</tag>` 

Important html `<tag>`s:

- `<html></html>`
- `<head></head>`
- `<body></body>`
- `<div></div>`
- `<p></p>`
- `<span></span>`
- `<section></section>`
- `<a href=""></a>`
- `<img src="">`
- `<br>`
- ```
<table>
    <thead>
        <tr>
            <th></th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <td></td>
            ...
        </tr>
    </tbody>
</table>
 ```
- `<ul></ul>` / `<ol></ol>` + `<li></li>`
    
Tags can have different attributes:
- `<a>`: href
- `<img>`: src
- id
- class
- anything that is not a html keyword

## Easy file sharing

Start your own web-server:
- in command prompt change your directory to `current/directory/data`
- start the server with the `python -m http.server` command

Open `localhost:8000/test.html` in your browser, and inspect the source of the page!
